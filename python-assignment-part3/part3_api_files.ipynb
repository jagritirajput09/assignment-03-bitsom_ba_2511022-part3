## Task 1 — File Read & Write Basics

```python
# Part A — Write
file_name = "python_notes.txt"

lines = [
    "Topic 1: Variables store data. Python is dynamically typed.\n",
    "Topic 2: Lists are ordered and mutable.\n",
    "Topic 3: Dictionaries store key-value pairs.\n",
    "Topic 4: Loops automate repetitive tasks.\n",
    "Topic 5: Exception handling prevents crashes.\n"
]

# Write mode
with open(file_name, "w", encoding="utf-8") as f:
    f.writelines(lines)
print("File written successfully.")

# Append mode
with open(file_name, "a", encoding="utf-8") as f:
    f.write("Topic 6: Functions improve code reusability.\n")
    f.write("Topic 7: APIs allow communication between systems.\n")
print("Lines appended.")

# Part B — Read
with open(file_name, "r", encoding="utf-8") as f:
    all_lines = f.readlines()

print("\nFile Contents:")
for i, line in enumerate(all_lines, start=1):
    print(f"{i}. {line.strip()}")

print("\nTotal lines:", len(all_lines))

# Keyword search
keyword = input("\nEnter keyword to search: ").lower()
found = False

for line in all_lines:
    if keyword in line.lower():
        print(line.strip())
        found = True

if not found:
    print("No matching lines found.")
```

## Task 2 — API Integration

```python
import requests

BASE_URL = "https://dummyjson.com/products"

# Safe request wrapper
def safe_get(url):
    try:
        response = requests.get(url, timeout=5)
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)
    return None

# Step 1 — Fetch products
res = safe_get(f"{BASE_URL}?limit=20")

if res and res.status_code == 200:
    data = res.json()["products"]

    print("ID | Title                     | Category     | Price   | Rating")
    print("---------------------------------------------------------------")

    for p in data:
        print(f"{p['id']:<3}| {p['title'][:25]:<25}| {p['category']:<12}| ${p['price']:<7}| {p['rating']}")

    # Step 2 — Filter & Sort
    filtered = [p for p in data if p["rating"] >= 4.5]
    filtered.sort(key=lambda x: x["price"], reverse=True)

    print("\nFiltered (Rating ≥ 4.5, Sorted by Price ↓):")
    for p in filtered:
        print(f"{p['title']} - ${p['price']} - Rating: {p['rating']}")

# Step 3 — Category search
res = safe_get(f"{BASE_URL}/category/laptops")

if res and res.status_code == 200:
    laptops = res.json()["products"]

    print("\nLaptops:")
    for p in laptops:
        print(f"{p['title']} - ${p['price']}")

# Step 4 — POST request
try:
    payload = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API"
    }

    res = requests.post(f"{BASE_URL}/add", json=payload, timeout=5)
    print("\nPOST Response:", res.json())

except requests.exceptions.ConnectionError:
    print("Connection failed.")
except requests.exceptions.Timeout:
    print("Timeout occurred.")
except Exception as e:
    print("Error:", e)
```

## Task 3 — Exception Handling

```python
import requests

# Part A — Safe divide
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))


# Part B — File reader
def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            return f.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")

print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))


# Part C — Robust API example
def safe_request(url):
    try:
        return requests.get(url, timeout=5)
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out.")
    except Exception as e:
        print("Unexpected error:", e)
    return None


# Part D — Input validation loop
while True:
    user_input = input("\nEnter product ID (1–100) or 'quit': ")

    if user_input.lower() == "quit":
        break

    if not user_input.isdigit():
        print("⚠ Invalid input. Enter a number.")
        continue

    pid = int(user_input)

    if not (1 <= pid <= 100):
        print("⚠ ID must be between 1 and 100.")
        continue

    res = safe_request(f"https://dummyjson.com/products/{pid}")

    if res:
        if res.status_code == 404:
            print("Product not found.")
        elif res.status_code == 200:
            product = res.json()
            print(f"{product['title']} - ${product['price']}")
```

## Task 4 — Logging to File

```python
from datetime import datetime
import requests

log_file = "error_log.txt"

def log_error(context, error_type, message):
    with open(log_file, "a", encoding="utf-8") as f:
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        f.write(f"[{timestamp}] ERROR in {context}: {error_type} — {message}\n")

# Trigger ConnectionError
try:
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError as e:
    print("Triggered ConnectionError")
    log_error("fetch_products", "ConnectionError", str(e))

# Trigger HTTP error manually
res = requests.get("https://dummyjson.com/products/999", timeout=5)

if res.status_code != 200:
    print("Triggered HTTP error (404)")
    log_error("lookup_product", "HTTPError", f"{res.status_code} Not Found for product ID 999")

# Read log file
print("\nError Log Contents:")
with open(log_file, "r", encoding="utf-8") as f:
    print(f.read())
```
